# Wrangle & Analyze WeRateDogs Data - Analysis & Insights Report

Upon completion of this projects, 3 insights were derived;
- Distribution of the rated dog by Stage
- The most liked and retweeted dog breed
- Periodic Engagement of the WeRateDog Twitter Account

## 1. Distribution of the rated dog by Stage

> Dogs in pupper stage were discovered to be the most rated, followed by those in doggo stage, then puppo. Dogs in floofer stage are rarely rated.

## 2.The most liked and retweeted dog breed

> First, the distribution of top 10 Most Tweeted Dog Breed was explored, Golden Retriever was found to be the most popular dog breed, followed by Labrador Retriever.

> Golden Retriever and Labrador Retriever turns out to be the dog breeds with most liked tweets

> Though Golden Retriever and Labrador Retriever lead the list of breeds with most liked tweets and retweet, these two are not among the Top 10 breed with highest average total tweets.
